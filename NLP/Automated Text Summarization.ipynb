```python
stopwords = list(STOP_WORDS)
```

```python
doc = nlp(text)
tokens = [token.text for token in doc]
tokens
```

```python
punctuation
```

```python
# Add new line to it
punctuation = punctuation + '\n'
punctuation
```

### Text Cleaning

```python
# If any key is being introduced for first time the occurrence is one
# If introduced second time, third time, it will be incremented by 1
word_frequencies = {}

for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
print(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

```python
# Normalizing by dividing by max_frequency
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word] / max_frequency
```

```python
print(word_frequencies)
```

### Sentence Tokenization 

```python
sentence_tokens = [sent for sent in doc.sents]
print(sentence_tokens)
```

```python
# Sentence Score
sentence_score = {}

for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_score.keys():
                sentence_score[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_score[sent] += word_frequencies[word.text.lower()]
```

```python
print(sentence_score)
```

```python
# Select 30% with maximum score
from heapq import nlargest
select_length = int(len(sentence_tokens)) * 0.3
```

```python
print(select_length)
```

### Getting Summary 

```python
summary = nlargest(n=int(select_length), iterable=sentence_score, key = sentence_score.get)
```

```python
print(summary)
```

```python
# Combine these sentences together
final_summary = [word.text for word in summary]
final_summary
```

```python
summary = ' '.join(final_summary)
print(summary)
```
